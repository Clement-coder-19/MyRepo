```python
import numpy as np
import matplotlib.pyplot as plot
import math

def sin_taylor(x, n_terms):
    s = 0
    for n in range(n_terms):
        s += ((-1)**n) * x**(2*n+1) / math.factorial(2*n+1)
    return s

def cos_taylor(x, n_terms):
    s = 0
    for n in range(n_terms):
        s += ((-1)**n) * x**(2*n) / math.factorial(2*n)
    return s

time = np.arange(-3, 3, 0.1)
n_terms = 2
amplitude_sin = np.sin(time)
amplitude_cos = np.cos(time)
sin_approx = [sin_taylor(t, n_terms) for t in time]
cos_approx = [cos_taylor(t, n_terms) for t in time]
plot.plot(time, amplitude_sin, color='blue', label="sin numpy")
plot.plot(time, sin_approx, color='orange', label="sin Taylor")
plot.title('Sine and Taylor approx')
plot.grid(True, which='both')
plot.ylim(-1.5, 1.5)
plot.legend()
plot.show()

plot.plot(time, amplitude_cos,color='blue', label="cos numpy")
plot.plot(time, cos_approx, color='orange', label="cos Taylor")
plot.title('Cos and Taylor approx')
plot.grid(True, which='both')
plot.ylim(-1.5, 1.5)
plot.legend()
plot.show()

error_sin = np.array(sin_approx) - amplitude_sin
error_cos = np.array(cos_approx) - amplitude_cos

plot.plot(time, error_sin)
plot.title('Diff of sine and Taylor approx')
plot.grid(True)
plot.ylim(-0.0100, 0.0100)
plot.show()


plot.plot(time, error_cos)
plot.title('Diff of cos and Taylor approx')
plot.grid(True)
plot.ylim(-0.0100, 0.0100)
plot.show()

formula = np.array(sin_approx)**2 + np.array(cos_approx)**2
plot.plot(time, formula, color='orange')
plot.title('sin^2(x) + cos^2(x)')
plot.axhline(y=1)
plot.ylim(0.5, 1.5)
plot.grid(True)
plot.show()

plot.plot(time, formula-1, color='blue')
plot.title('sin^2(x) + cos^2(x)')
plot.ylim(-0.5, 0.5)
plot.grid(True)
plot.show()
```
